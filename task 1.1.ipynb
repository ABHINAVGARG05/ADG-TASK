```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.subplot(2,2,1)
sns.histplot(songs['Title'])
```

```python
sns.histplot(songs['Popularity'],bins = 20, kde =True)
plt.title('Popularity Distribution')
plt.xlabel('Popularity')
plt.ylabel('Count')
plt.show()
```

```python

sns.lineplot(x='ReleaseDate', y='Popularity', data=songs)
plt.title('Popularity Over Time')
plt.xlabel('Release Date')
plt.ylabel('Popularity')
```

```python
sns.histplot(x='Genre', y='Popularity', data=songs)
plt.title('Popularity by Genre')
plt.xlabel('Genre')
plt.ylabel('Popularity')
```

```python
top_artists = songs['Artist'].value_counts().head(10).index
```

```python
sns.boxplot(x='Artist', y='Popularity', data=songs[songs['Artist'].isin(top_artists)])
plt.title('Popularity by Top 10 Artists')
plt.xlabel('Artist')
plt.ylabel('Popularity')
```
